```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

import module_heliquest_like
from module_heliquest_like import amino_acids
```

```python
df = pd.read_csv('./Output/NEES_summary_with_heliquest_Median_Transformed.csv')
```

```python
# Select only those with categories 1, 2, 3
df = df[(df['Category'] == 1) | (df['Category'] == 2) | (df['Category'] == 3)]
```

```python
df_selected = df.copy()[selected_columns]
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(df_selected, annot=False, cmap='plasma', fmt=".2f", robust=True)
plt.savefig('./Output/All.pdf', format='pdf', bbox_inches='tight')
plt.show()
plt.close()
```
