```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('./Output/NEES_summary_with_heliquest_Median_Transformed.csv')
```

```python
plt.figure(figsize=(10, 6))
scatter = plt.scatter(df['HydrophobicFaceWidth'], df['phil_S_T'], c=df['Median_boxcox'], cmap='viridis', alpha=0.7, marker='o')  # Use 'o' for filled circles

# Annotate points with their respective 'AH' values
for i, txt in enumerate(df['AH']):
    plt.annotate(txt, (df['HydrophobicFaceWidth'][i], df['phil_S_T'][i]), fontsize=8, alpha=0.6)

plt.colorbar(scatter, label='Median Boxcox Value')
plt.xlabel('Hydrophobic Face Width')
plt.ylabel('phil_S_T')
plt.title('Hydrophobic Face Width vs phil_S_T colored by Median Boxcox Value with AH labels')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python

x = 'HydrophobicFaceWidth'
y = 'Netcharge'
z = 'phil_S_T'
```

```python
# Create the 3D scatter plot with all axes starting from zero
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')


# Set all axes to start from zero
ax.set_xlim(0, df[x].max() + 1)
ax.set_ylim(0, df[y].max() + 1)
ax.set_zlim(0, df[z].max())

scatter = ax.scatter(df[x], df[y], df[z], c=df['Median_boxcox'], cmap='viridis', alpha=0.7, marker='o')

# Annotate points with their respective 'AH' values
for i, txt in enumerate(df['AH']):
    ax.text(df[x][i], df[y][i], df[z][i], txt, fontsize=8, alpha=0.6)

plt.colorbar(scatter, label='Median Boxcox Value')
ax.set_xlabel(x)
ax.set_ylabel(y)
ax.set_zlabel(z)
plt.title('3D Scatter Plot of Hydrophobic Face Width, ST:RK ratio, and Netcharge colored by Median Boxcox Value with AH labels (All Axes Starting from Zero)')
plt.show()
```
